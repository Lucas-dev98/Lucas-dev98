```python
import datetime

class Livro:
    def __init__(self, titulo, editora, quantidade, autor, status='disponivel'):
        self.titulo = titulo
        self.editora = editora
        self.quantidade = quantidade
        self.autor = autor
        self.status = status
        self.historico = PilhaHistorico()

    def emprestar(self):
        if self.quantidade > 0 and self.status == 'disponivel':
            self.quantidade -= 1
            self.status = 'emprestado'
            return True
        else:
            return False

    def devolver(self):
        self.quantidade += 1
        self.status = 'disponivel'

class Historico:
    def __init__(self, acao, data):
        self.acao = acao
        self.data = data

class PilhaHistorico:
    def __init__(self):
        self.historico = []

    def adicionar_historico(self, acao):
        now = datetime.datetime.now()
        self.historico.append(Historico(acao, now))

class ListaLivros:
    def __init__(self):
        self.livros = []

    def adicionar_livro(self, livro):
        self.livros.append(livro)

    def buscar_livro(self, titulo):
        for livro in self.livros:
            if livro.titulo == titulo:
                return livro
        return None

class FilaEspera:
    def __init__(self):
        self.fila = []

    def adicionar_fila(self, usuario):
        self.fila.append(usuario)

    def remover_fila(self):
        if self.fila:
            return self.fila.pop(0)
        else:
            return None

def mensagem_disponibilidade(livro):
    if livro.status == 'disponivel':
        return f'O livro "{livro.titulo}" está disponível para empréstimo! Quantidade: {livro.quantidade}'
    else:
        return f'O livro "{livro.titulo}" foi devolvido e está disponível para empréstimo. Quantidade: {livro.quantidade}'

def mostrar_menu(lista_livros):
    print("\n### Menu ###")
    print("1. Listar Livros Disponíveis")
    print("2. Pegar Livro Emprestado")
    print("3. Devolver Livro")
    print("4. Cadastrar Novo Livro")
    print("5. Sair do Sistema")
    print("############\n")

    print("### Livros Disponíveis ###")
    for livro in lista_livros.livros:
        print(f"{livro.titulo} - Autor: {livro.autor} - Editora: {livro.editora} - Quantidade: {livro.quantidade}")
    print("###########################")

def pegar_livro_emprestado(lista_livros, fila_espera):
    titulo = input("Digite o título do livro que deseja pegar emprestado: ")
    livro = lista_livros.buscar_livro(titulo)
    if livro:
        if livro.emprestar():
            print(f"Você pegou o livro '{livro.titulo}' emprestado!")
        else:
            print("Desculpe, este livro não está disponível no momento.")
            fila_espera.adicionar_fila(input("Digite seu nome para entrar na fila de espera: "))
    else:
        print("Livro não encontrado.")

def devolver_livro(lista_livros):
    titulo = input("Digite o título do livro que deseja devolver: ")
    livro = lista_livros.buscar_livro(titulo)
    if livro:
        livro.devolver()
        print(f"O livro '{livro.titulo}' foi devolvido.")
    else:
        print("Livro não encontrado.")

def cadastrar_novo_livro(lista_livros):
    titulo = input("Digite o título do novo livro: ")
    autor = input("Digite o nome do autor: ")
    editora = input("Digite o nome da editora: ")
    quantidade = int(input("Digite a quantidade de cópias: "))
    novo_livro = Livro(titulo, editora, quantidade, autor)
    lista_livros.adicionar_livro(novo_livro)
    print(f"O livro '{titulo}' foi cadastrado com sucesso!")

# Criando alguns livros para o exemplo
livro1 = Livro("Python for Beginners", "Editora A", 5, "John Smith")
livro2 = Livro("Java Programming", "Editora B", 3, "Emily Johnson")
livro3 = Livro("Machine Learning", "Editora C", 2, "Alice Lee")

lista_livros = ListaLivros()
lista_livros.adicionar_livro(livro1)
lista_livros.adicionar_livro(livro2)
lista_livros.adicionar_livro(livro3)

fila_espera = FilaEspera()

while True:
    mostrar_menu(lista_livros)
    opcao = input("Digite o número da opção desejada: ")

    if opcao == '1':
        # A lista de livros disponíveis já é mostrada no menu
        pass
    elif opcao == '2':
        pegar_livro_emprestado(lista_livros, fila_espera)
    elif opcao == '3':
        devolver_livro(lista_livros)
    elif opcao == '4':
        cadastrar_novo_livro(lista_livros)
    elif opcao == '5':
        print("Saindo do sistema. Até logo!")
        break
    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")
```
